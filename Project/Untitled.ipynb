```python
data=pd.read_csv('bib_final_utf.csv')
```

```python
data=data.drop('Unnamed: 0',axis=1)
data=data.drop_duplicates()
data=data.dropna()
```

```python
data.head(10)
```

```python
with open('libraries.ttl','w') as f:
    f.writelines('@prefix ex: <http://sample.com/> .\n')
    f.writelines('@prefix prop: <http://sample.com/prop/> .\n')
    f.writelines('@prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#> .\n')
    f.writelines('@prefix type: <http://sample.com/type/> .\n')
    f.writelines('\n')
    for i in range(data.shape[0]):
        f.write(f'ex:Lib{i+1} rdf:type type:Library .\n')
    f.write('\n')
    for i in range(data.shape[0]):
        f.write(f'<http://sample.com/Lib{i+1}>\n')
        f.write(f"\t<http://sample.com/prop/Name> \"{str(data.iloc[i]['Libelle1']).lower()}\" ;\n")
        f.write(f"\t<http://sample.com/prop/Address> \"{str(data.iloc[i]['voie']).lower()}\" ;\n")
        f.write(f"\t<http://sample.com/prop/Zipcode> \"{int(data.iloc[i]['CP'])}\" ;\n")
        if len(str(int(data.iloc[i]['CP'])))==5:
            f.write(f"\t<http://sample.com/prop/Dpt> \"{str(int(data.iloc[i]['CP']))[:2]}\" ;\n")
        else:
            f.write(f"\t<http://sample.com/prop/Dpt> \"{str(int(data.iloc[i]['CP']))[0]}\" ;\n")
        f.write(f"\t<http://sample.com/prop/City> \"{str(data.iloc[i]['Ville']).capitalize()}\" ;\n")
        f.write(f"\t<http://sample.com/prop/Surface> \"{data.iloc[i]['surface']}\" ;\n")
        f.write(f"\t<http://sample.com/prop/Lattitude> \"{data.iloc[i]['lat']}\" ;\n")
        f.write(f"\t<http://sample.com/prop/Longitude> \"{data.iloc[i]['long']}\" .\n")
        f.write('\n')
    
```
